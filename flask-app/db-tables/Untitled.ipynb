## Now first using Kmeans 

```python
# Apply K-means with different n_init values and choose the one with the highest silhouette score
best_score = -1
best_kmeans = None

for n_init in range(10, 21):
    kmeans = KMeans(n_clusters=3, init='k-means++', n_init=n_init, random_state=42)
    kmeans.fit(hosp)
    labels = kmeans.labels_
    score = silhouette_score(hosp, labels)
    if score > best_score:
        best_score = score
        best_n_init = n_init
        best_kmeans = kmeans

labels = kmeans.labels_
print(labels)
```

```python
print("Best n_init:", best_n_init)
print("Best Silhouette Score:", best_score)
```

```python
labels.max()
```

## Now using fuzzy c means

```python
import skfuzzy as fuzz
n_clusters = 3
# Specify the fuzziness parameter (typically set to 2 for FCM)
m = 2
# Perform Fuzzy C-means clustering
cntr, u, _, _, _, _, _ = fuzz.cluster.cmeans(hosp.T, n_clusters, m, error=0.005, maxiter=1000)

# 'u' contains the membership degrees of each hospital for each cluster

# 'cntr' contains the final cluster centers

# Now, you can use the membership degrees to assign hospitals to risk categories
risk_categories = np.argmax(u, axis=0)  # Assign each hospital to the cluster with the highest m
fuzzy_silhouette_score = silhouette_score(hosp, risk_categories)
print("Fuzzy C-means Silhouette Score:", fuzzy_silhouette_score)
```

## Now using Agglomerative Hierarchical Clustering

```python
from sklearn.cluster import AgglomerativeClustering

# Perform Agglomerative Hierarchical Clustering with 3 clusters
n_clusters = 3
agg_clustering = AgglomerativeClustering(n_clusters=n_clusters)
agg_labels = agg_clustering.fit_predict(hosp)

agg_silhouette_score = silhouette_score(hosp, agg_labels)

print("Agglomerative Clustering Labels:", agg_labels)
print("Agglomerative Clustering Silhouette Score:", agg_silhouette_score)

```

## Now using GMM

```python
from sklearn.mixture import GaussianMixture

# Perform Gaussian Mixture Model clustering
n_components = 3  # The number of clusters you desire
gmm = GaussianMixture(n_components=n_components, random_state=42)
gmm.fit(hosp)

gmm_labels = gmm.predict(hosp)

gmm_silhouette_score = silhouette_score(hosp, gmm_labels)

print("GMM Labels:", gmm_labels)
print("GMM Silhouette Score:", gmm_silhouette_score)
```

```python
import random
import pymongo
from pymongo import MongoClient
from bson import Binary

# Connect to the MongoDB server
client = MongoClient('mongodb://localhost:27017/')
db = client['ey_hosp']
collection = db['documents']

# List of 5 sample reports with their file paths
sample_reports = [
    {'document_name': 'RFT1', 'file_path': 'sample reports/RFT1.png'},
    {'document_name': 'RFT2', 'file_path': 'sample reports/RFT2.png'},
    {'document_name': 'RFT3', 'file_path': 'sample reports/RFT3.png'},
]

# Generate and insert documents
for patient_id in range(1, 101):  # Assuming patient IDs from 1 to 100
    random_report = random.choice(sample_reports)
    with open(random_report['file_path'], 'rb') as file:
        binary_data = Binary(file.read())
    
    document = {
        'patient_id': patient_id,
        'document_name': random_report['document_name'],
        'image_binary': binary_data
    }
    collection.insert_one(document)

print("Documents inserted successfully.")
```

```python
import pandas as pd
import numpy as np
from faker import Faker
import random
from datetime import datetime, timedelta

# Create fake data generator
fake = Faker()

# Create a list to store patient data
patients = []

# List of treatment codes and their explanations
treatment_codes = {
    'NEP': 'Nephrology',
    'PED': 'Pediatric',
    'CAT': 'Cataract',
    'OPT': 'Ophthalmology',
    'ORT': 'Orthopedic',
    'CAR': 'Cardiology',
    'GEN': 'General',
    'ENT': 'ENT (Ear, Nose, Throat)',
    'DER': 'Dermatology',
    'OTH': 'Other'
}

sample_prescriptions = [
    "Take 1 tablet of medicine A every morning for 7 days.",
    "Apply ointment B to the affected area twice a day for 10 days.",
    "Take 2 capsules of medicine C after meals for 5 days.",
    "Use eye drops D in both eyes every 4 hours as needed.",
    "Take 1 tablet of medicine E before bedtime for 3 days."
]

# Generate patient data with multiple visits
for _ in range(100):
    patient_id = fake.unique.random_int(min=1000, max=9999)
    full_name = fake.name()
    gender = random.choice(['Male', 'Female'])
    patient_type = random.choice(['Inpatient', 'Outpatient'])
    treatment_code = random.choice(list(treatment_codes.keys()))
    action_taken = random.choice(['Consulted', 'Operated', 'Tested'])
    next_visit_suggestion = random.choice(['Prescription', 'Operation', 'Test'])
    prescription = random.choice(sample_prescriptions)

    # Generate visit dates in chronological order
    visit_dates = [fake.date_time_this_year() for _ in range(random.randint(1, 5))]
    visit_dates.sort()

    # Add data for each visit
    for visit_date in visit_dates:
        patients.append([
            patient_id, full_name, gender, patient_type, treatment_code,
            action_taken, next_visit_suggestion, prescription, visit_date
        ])

# Create a pandas DataFrame from the generated data
data_columns = [
    'PatientID', 'FullName', 'Gender', 'PatientType', 'TreatmentCode',
    'ActionTaken', 'NextVisitSuggestion', 'Prescription', 'VisitDate'
]
patients_df = pd.DataFrame(patients, columns=data_columns)
patients_df
```

```python

# Save the DataFrame to a CSV file
patients_df.to_csv('patients_data.csv', index=False)

print("Data generation and CSV file creation completed.")
```

### Pgeocodes is a python library for high performance quesring of GPS coordinates, region name and municipality name from postal codes. Distances between postal codes as well as general distance queries are also supported. To use this file of your country must be downloaded from geonames database. currently distance query of the postal codes in the same country is supported.

```python
import pgeocode

dist = pgeocode.GeoDistance('IN')
dist.query_postal_code('110034', '110056')
```

```python
import pandas as pd
df = pd.read_fwf('IN.txt',header = None)
df
```

```python
df1 = df[[1]].sample(frac=1).reset_index(drop=True)
df1
#here sample(frac = 1) is used to shuffle the datdframe and reset_index is used to make sure that the index does 
#get shuffled and starts from 0 only. drop = True is used to make sure that pandas does not make a coloumn for the previous index
```

```python
df2 = df1.head(100)
col = ['index 1']
df2.columns = col
df2
```

```python
df3= df1.iloc[100:200].reset_index(drop = True)
columns = ['index 2']
df3.columns = columns
df3
```

```python
df4 = pd.concat([df2, df3], axis=1)
df4
```

```python
df4.dtypes
```

```python
df4['index 1'] = df4['index 1'].astype("str")
df4['index 2'] = df4['index 2'].astype("str")
df4.dtypes
```

```python
df4
```

```python
df4['index 2'][0]
```

```python
dist = pgeocode.GeoDistance('IN')
df4['dist'] = dist.query_postal_code(df4['index 1'].to_list(), df4['index 2'].to_list())
df4
```

```python
import geopy.distance
import pandas as pd
def calc_distance(x):
    site_coords = (12.048367, 79.572156)
    place2_coords = (x.latitude, x.longitude)
    return geopy.distance.geodesic(site_coords, place2_coords ).km
df5['distance'] = df5.apply(calc_distance, axis = 1)
# here i have calculated the distance of all the lat and lon from a single point.
     
```
